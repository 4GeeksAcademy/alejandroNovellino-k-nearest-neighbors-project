# Create the third table

```python
# query to make the third table
query = f'''
    SELECT movie.title, movie.overview, movie.genres, movie.keywords, credit.movie_id, credit.cast, credit.crew
    FROM movie, credit
    WHERE movie.title = credit.title
'''

# dataframe to save the data
third_table = pd.DataFrame()

with engine.connect() as conn:
    # third column as df
    third_table = pd.read_sql(query, conn)
```

```python
# verify that the third table was selected successfully
third_table.head()
```

## Save it to the database

```python
with engine.connect() as conn:
    # dave the table to the database
    third_table.to_sql(name='third_table', con=engine, if_exists='replace')
```

# Process the table

Transform each json column to just the needed data

```python
import ast

# process the genres column
third_table['genres'] = third_table['genres'].apply(lambda json_element: [element['name'] for element in ast.literal_eval(json_element)])
# process the keywords column
third_table['keywords'] = third_table['keywords'].apply(lambda json_element: [element['name'] for element in ast.literal_eval(json_element)])
# process the cast column
third_table['cast'] = third_table['cast'].apply(lambda json_element: [element['name'] for element in ast.literal_eval(json_element)[:3]])
# process the crew column
third_table['crew'] = third_table['crew'].apply(lambda json_element: [element['name'] for element in ast.literal_eval(json_element)[:3] if element['job'] == 'Director'])
```

```python
third_table.head()
```

Now join each list column

```python
# process the genres column
third_table['genres'] = third_table['genres'].apply(lambda list_element: " ".join([element.replace(' ', '') for element in list_element]))
# process the cast column
third_table['cast'] = third_table['cast'].apply(lambda list_element: " ".join([element.replace(' ', '') for element in list_element]))
# process the crew column
third_table['crew'] = third_table['crew'].apply(lambda list_element: " ".join([element.replace(' ', '') for element in list_element]))
# process the crew column
third_table['keywords'] = third_table['keywords'].apply(lambda list_element: " ".join([element.replace(' ', '') for element in list_element]))
```

Add all the columns to just one named tags

```python
third_table['tags'] = third_table[['overview', 'genres', 'cast', 'crew', 'keywords']].apply(lambda row: " ".join([row['overview'], row['genres'], row['keywords'], row['cast'], row['crew']]), axis=1)
```
